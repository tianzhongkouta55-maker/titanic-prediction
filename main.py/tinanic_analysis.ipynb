```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. データの読み込み
df = pd.read_csv('C:\\Users\\User\\Downloads\\titanic-prediction\\data\\train.csv')

# 2. 前処理：文字の数値化と欠損値の補完
df['Sex'] = df['Sex'].map({'female': 0, 'male': 1})
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

# 特徴量（すべての数値データ）とラベルを選択
X = df[['Sex', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']]
y = df['Survived']

# 3. データの分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 4. モデル作成：あえて制限なし（過学習の状態を作る）
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)

# 5. 正解率の算出：訓練データとテストデータの差を確認
train_acc = model.score(X_train, y_train) * 100
test_acc = model.score(X_test, y_test) * 100

print(f"訓練データ正解率: {train_acc:.2f}%") # 100%に近いほど「丸暗記」している
print(f"テストデータ正解率: {test_acc:.2f}%") # 訓練データとの差が大きいほど「過学習」
print("-" * 30)

# 6. 特徴量の重要度：どのデータが判断に寄与したか
importances = model.feature_importances_
for name, imp in zip(X.columns, importances):
    print(f"{name}: {imp:.4f}")

# 重要度の可視化
plt.figure(figsize=(8, 4))
plt.barh(X.columns, importances)
plt.title("Feature Importance")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. データの読み込み
df = pd.read_csv('C:\\Users\\User\\Downloads\\titanic-prediction\\data\\train.csv')

# 2. 前処理：文字の数値化と欠損値の補完
df['Sex'] = df['Sex'].map({'female': 0, 'male': 1})
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

# 新しい特徴量「FamilySize」を作成（自分＋兄弟配偶者＋親子）
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# 特徴量（すべての数値データ）とラベルを選択
X = df[['Sex', 'Pclass', 'Age', 'Fare', 'FamilySize']]
y = df['Survived']

# 3. データの分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 4. 木の深さを「3」に制限してみる（＝細かいことは無視させる）
model = DecisionTreeClassifier(max_depth=3, random_state=0)
model.fit(X_train, y_train)

# 5. 正解率の算出：訓練データとテストデータの差を確認
train_acc = model.score(X_train, y_train) * 100
test_acc = model.score(X_test, y_test) * 100

print(f"訓練データ正解率: {train_acc:.2f}%") # 100%に近いほど「丸暗記」している
print(f"テストデータ正解率: {test_acc:.2f}%") # 訓練データとの差が大きいほど「過学習」
print("-" * 30)

# 6. 特徴量の重要度：どのデータが判断に寄与したか
importances = model.feature_importances_
for name, imp in zip(X.columns, importances):
    print(f"{name}: {imp:.4f}")

# 重要度の可視化
plt.figure(figsize=(8, 4))
plt.barh(X.columns, importances)
plt.title("Feature Importance")
plt.show()
```
