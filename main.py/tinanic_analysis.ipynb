```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. データの読み込み
df = pd.read_csv('C:\\Users\\User\\Downloads\\titanic-prediction\\data\\train.csv')

# 2. 使うデータを選び、文字を数字に変換する（機械は数字しか読めないため）
df['Sex'] = df['Sex'].map({'female': 0, 'male': 1}) # 女性を0、男性を1に変換.

# 特徴量（ヒントとなるデータ）とラベル（正解）に分ける
X = df[['Sex', 'Pclass']] # 性別と客室ランクを使う
y = df['Survived']        # 生存したかどうか

# 3. データを「学習用」と「テスト用」に分ける
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)#stratiify=yを追加すると正答率が下がるため今回は追加していない．

# 4. モデル（AIの脳）を作って学習させる
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# 5. 予測して、正解率を出す
predictions = model.predict(X_test)
print(f"正解率: {accuracy_score(y_test, predictions) * 100:.2f}%")
```
